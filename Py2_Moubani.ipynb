# 6.Histogram of total bill column


```python
#Histogram on total_bill column

plt.figure(figsize=(10,4))
plt.hist(df['total_bill'],bins=20,color='beige',edgecolor='brown',rwidth=0.8)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid('True')
plt.title('Histogram of Total Bill')
plt.show()
```

Interpretation:The histogram created visualizes the distribution of the total_bill values. The x-axis represents the different bins or ranges of total bill amounts, and the y-axis represents the frequency or count of bills falling within each bin.

```python
#KDE plot on total_plot

plt.figure(figsize=(10,4))
sns.kdeplot(data=df['total_bill'],fill=True,color='indigo',edgecolor='pink')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid('True')
plt.title('KDE Plot of Total Bill')
plt.show()
```

Interpretation:The KDE plot generated is a smoothed representation of the distribution.The higher the curve, the denser the data points are at that value.

In the plot,the peak of the curve is around 15-20, indicating that this is the most common range of total bills.

The filled area under the curve represents the probability of observing a total bill within a given range.

# 7.Box plot for total_bill column

```python
#Box plot for total_bill column
plt.figure(figsize=(8,4))
sns.boxplot(df['total_bill'],color='pink',orient='h')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Box Plot for total_bill column')
plt.show()
```

Interpretation:The box plot generated shows the following:
Median: The line inside the box represents the median of the data. In your plot, the median is around 17.

Interquartile Range (IQR): The box itself represents the interquartile range, which contains the middle 50% of the data. The bottom of the box is the 25th percentile (Q1), and the top is the 75th percentile (Q3).

Outliers: Data points outside the whiskers are considered outliers and are plotted as individual points.

# 9.Correlation matrix of the dataset.

```python
#Calculating and visualizing the correlation matrix of the dataset
corr_matrix = df.select_dtypes(include=np.number).corr()
print(corr_matrix)
plt.figure(figsize=(8,4))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Heat Map to visualise the Correlation matrix of Tips Dataset')
plt.show()
```

Interpretation:The correlation matrix generated shows the pairwise correlations between the numerical variables in your dataset.

Here,there is a strong positive correlation between total_bill and tip.And a weak positive correlation between tip and size.

#10. t-test to compare the means of 'total_bill' between smokers and non-smokers


```python
# Perform t-test
from scipy.stats import ttest_ind
smoker_bill = df[df['smoker'] == 'Yes']['total_bill']
nonsmoker_bill = df[df['smoker'] == 'No']['total_bill']
t_stat, p_value = ttest_ind(smoker_bill, nonsmoker_bill)

# Print results
print(f'T-statistic: {t_stat:.2f}')
print(f'P-value: {p_value:.3f}')
```

The t-statistic of 2.99 suggests that there is a noticeable difference between the means of the two groups.

The p-value of 0.003, being less than the common significance level of 0.05, indicates that the difference in means is statistically significant.
